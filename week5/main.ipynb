## Q1
### 최소 비용 경로 (단방향, 가중치)
#### BOJ 1916 최소비용 구하기 (유형 참고)

#### 문제
각 도시는 버스(방향 간선)로 연결되고 노선마다 비용이 다릅니다. 출발 도시에서 도착 도시까지 드는 **최소 비용**을 구합니다. 같은 도시 쌍 `(u, v)`에 여러 노선이 있을 수 있으며, 비용은 양의 정수입니다.

#### 입력
- 첫 줄: 도시 수 `N` (`1 ≤ N ≤ 1,000`)
- 둘째 줄: 버스 수 `M` (`1 ≤ M ≤ 100,000`)
- 다음 `M`줄: `u v w` — `u`에서 `v`로 가는 비용 `w` (단방향)
- 마지막 줄: 출발 도시, 도착 도시

#### 출력
- 출발에서 도착까지 최소 비용 (경로 존재 보장)

#### 제한
- `w ≤ 100,000`, 시간 0.5초, 메모리 128MB

#### 예제 입력 1
```text
5
8
1 2 2
1 3 3
1 4 1
1 5 10
2 4 2
3 4 1
3 5 1
4 5 3
1 5
```

#### 예제 출력 1
```text
4
```

````python
INF = int(1e9)  # 무한을 의미하는 값 (10억) - 초기 거리로 사용

# n : int (정점의 개수)
# graph : list[list[tuple[int, int]]] (인접리스트 구조의 그래프), 튜플 : (v,cost) (도착 정점 번호, 가중치)
# start : int (시작 정점 번호)
def dijkstra(n, graph, start):
    dist = [INF] * (n + 1)      # 시작점에서 각 정점까지의 최단 거리 리스트
    visited = [False] * (n + 1) # 각 정점의 방문 여부
    dist[start] = 0             # 시작점은 거리 0으로 설정
    for _ in range(n): # 모든 정점을 최대 1번씩 처리
        min_dist = INF
        u = -1
        # 아직 방문하지 않은 정점 중에서 가장 dist가 작은 정점 u 선택
        for i in range(1, n + 1):
            if not visited[i] and dist[i] < min_dist:
                min_dist = dist[i]
                u = i
        if u == -1: # 연결되지 않은 나머지 정점만 남았다면 중단
            break
        visited[u] = True # u는 이제 최단 거리 확정됨
        # 선택된 정점 u와 인접한 정점들에 대해 거리 갱신 시도
        for v, cost in graph[u]:
            if dist[v] > dist[u] + cost:
                dist[v] = dist[u] + cost
    return dist # 최단 거리 리스트 반환


# 입력 처리
n = int(input())  # 도시 수 (정점 개수)
m = int(input())  # 버스 수 (간선 개수)

graph = [[] for _ in range(n + 1)]  # 인접 리스트 초기화

for _ in range(m):
    u, v, w = map(int, input().split())
    graph[u].append((v, w))  # u에서 v로 가는 비용 w인 간선 저장 (단방향)

start, end = map(int, input().split())  # 시작 도시와 도착 도시 입력
dist = dijkstra(n, graph, start)        # 다익스트라 실행
print(dist[end])                        # 시작 → 도착 도시까지의 최소 비용 출력
````

### 핵심 아이디어
- **가중치가 모두 양수**이므로, 한 번 확정된 최단 거리는 이후에 더 짧아지지 않습니다. 이 성질을 이용한 알고리즘이 **다익스트라**입니다.
- 매 단계에서 아직 확정되지 않은 정점 중 `dist`가 가장 작은 정점을 고르면, 그 정점의 거리는 최단으로 확정됩니다(음의 간선이 없을 때).
- 확정된 정점 `u`에서 나가는 간선 `(u, v, w)`에 대해 `dist[v] = min(dist[v], dist[u] + w)`로 갱신합니다.
- 구현은 **인접 리스트**에 `(도착, 비용)`을 저장하고, 우선순위 큐 없이 정점 수만큼 선형 탐색으로 최소 `dist` 정점을 찾는 **O(V²)** 버전입니다. `N ≤ 1,000`이면 충분합니다.

```mermaid
graph TD
    A[시작 정점 dist 0] --> B[미방문 중 dist 최소 정점 u 선택]
    B --> C[u 방문 처리 및 확정]
    C --> D[u의 이웃 v에 대해 dist 갱신]
    D --> B
    B --> E[모든 정점 처리 또는 더 이상 갱신 불가]
    E --> F[dist 도착점 출력]
```

### 알고리즘 흐름도
```mermaid
flowchart TD
    S[시작] --> I[dist INF, dist start 0, visited 전부 False]
    I --> L{n번 반복}
    L --> P[미방문 i 중 dist 최소인 u 선택]
    P --> Q{u가 없음?}
    Q -->|예| Z[종료]
    Q -->|아니오| V[visited u True]
    V --> R[u의 각 이웃 v, cost]
    R --> U{dist v 더 큼?}
    U -->|예| W[dist v = dist u + cost]
    U -->|아니오| L
    W --> L
    L --> Z
    Z --> O[dist end 출력]
```

### 예제 추적
출발 `1`, 도착 `5`. `1 → 4`(비용 1) 후 `4 → 5`(비용 3)이면 합계 **4**이고, `1 → 5` 직행(10)보다 저렴합니다.

```mermaid
graph LR
    A[1] -->|1| B[4]
    B -->|3| C[5]
    A -.->|10| C
```

## Q2
### 1번에서 N번까지, 두 필수 경유지
#### BOJ 1504 특정한 최단 경로 (유형 참고)

#### 문제
`1`번에서 `N`번으로 가되, 정점 `v1`과 `v2`를 **둘 다** 한 번씩 거쳐야 합니다. **들르는 순서는 자유**입니다 (`1 → v1 → v2 → N` 또는 `1 → v2 → v1 → N`). 최소 비용을 구하고, 불가능하면 `-1`을 출력합니다.

#### 입력
- 첫 줄: `N M` (`2 ≤ N ≤ 800`, `0 ≤ M ≤ 200,000`)
- 다음 `M`줄: `a b c` — `a`와 `b`를 **양방향**으로 잇는 비용 `c` (정점 쌍당 간선 최대 1개)
- 마지막 줄: `v1 v2` (`v1 ≠ v2`, `v1 ≠ N`, `v2 ≠ 1`)

#### 출력
- 조건을 만족하는 최소 비용, 없으면 `-1`

#### 제한
- `1 ≤ c ≤ 1,000`, 시간 1초, 메모리 256MB

#### 예제 입력 1
```text
4 6
1 2 3
2 3 3
3 4 1
1 3 5
2 4 5
1 4 4
2 3
```

#### 예제 출력 1
```text
7
```

````python
INF = int(1e9)

def dijkstra(n, graph, start):
    dist = [INF] * (n + 1)
    visited = [False] * (n + 1)
    dist[start] = 0

    for _ in range(n):
        min_dist = INF
        u = -1
        for i in range(1, n + 1):
            if not visited[i] and dist[i] < min_dist:
                min_dist = dist[i]
                u = i
        if u == -1:
            break

        visited[u] = True
        for v, cost in graph[u]:
            if dist[v] > dist[u] + cost:
                dist[v] = dist[u] + cost

    return dist

# 입력
n, m = map(int, input().split())
graph = [[] for _ in range(n + 1)]
for _ in range(m):
    a, b, c = map(int, input().split())
    # 양방향
    graph[a].append((b, c))
    graph[b].append((a, c))

v1, v2 = map(int, input().split())

# 다익스트라 3번만 호출
dist_from_1  = dijkstra(n, graph, 1)
dist_from_v1 = dijkstra(n, graph, v1)
dist_from_v2 = dijkstra(n, graph, v2)

# 1 -> v1 -> v2 -> n
path1 = dist_from_1[v1] + dist_from_v1[v2] + dist_from_v2[n]
# 1 -> v2 -> v1 -> n
path2 = dist_from_1[v2] + dist_from_v2[v1] + dist_from_v1[n]

ans = min(path1, path2)
print(-1 if ans >= INF else ans)
````

### 핵심 아이디어
- 전체 경로는 두 가지 패턴뿐입니다: **`1 → v1 → v2 → N`** 과 **`1 → v2 → v1 → N`**. 순서만 바뀐 동일한 구조입니다.
- 각 구간은 모두 **양의 가중치**이므로 구간별로도 최단 경로를 쪼개서 이어붙이면 됩니다(최적 부분 구조). 따라서 필요한 것은 다음 최단거리들입니다.
  - `1`에서 출발하는 최단거리, `v1`에서 출발하는 최단거리, `v2`에서 출발하는 최단거리.
- **다익스트라를 시작점 `1`, `v1`, `v2`에 대해 세 번**만 실행하면, 위 구간 길이를 모두 `O(1)`로 조합할 수 있습니다.
- `path1 = d(1,v1) + d(v1,v2) + d(v2,N)`, `path2 = d(1,v2) + d(v2,v1) + d(v1,N)`의 최소를 취하고, 어느 한 구간이라도 도달 불가(`≥ INF`)이면 `-1`입니다.

```mermaid
graph TD
    A[다익스트라 1, v1, v2에서 각각 실행] --> B[path1과 path2 비용 계산]
    B --> C[둘 중 최소]
    C --> D{INF 이상?}
    D -->|예| E[-1]
    D -->|아니오| F[최소 비용 출력]
```

### 알고리즘 흐름도
```mermaid
flowchart TD
    S[시작] --> G[양방향 그래프 입력]
    G --> D1[다익스트라 시작점 1]
    D1 --> D2[다익스트라 시작점 v1]
    D2 --> D3[다익스트라 시작점 v2]
    D3 --> P1[path1 합산]
    P1 --> P2[path2 합산]
    P2 --> M[min path1 path2]
    M --> C{M이 INF 이상?}
    C -->|예| N1[-1 출력]
    C -->|아니오| Y[M 출력]
```
