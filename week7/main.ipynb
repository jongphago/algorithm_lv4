## Q1
### [도전 기출문제] 디랩이와 관계 거리 합
#### BOJ 1389 케빈 베이컨의 6단계 법칙 (동일 유형)

#### 문제
학원에 학생 `N`명이 있고, 친구 관계는 양방향이다. 두 학생 사이의 관계 거리는 최단 경로의 길이(간선 수)로 정의한다. 직접 친구면 거리 1, 친구의 친구면 거리 2와 같다.

각 학생 `i`에 대해, 다른 모든 학생까지의 관계 거리의 합을 구한다. 이 합이 가장 작은 학생의 번호를 출력한다. 합이 같으면 번호가 더 작은 학생을 선택한다.

#### 입력
- 첫 줄: 학생 수 `N`, 친구 관계 수 `M` (`2 <= N <= 100`, `1 <= M <= 5000`)
- 다음 `M`줄: 친구 쌍 `A B` (`1 <= A, B <= N`, `A != B`, 양방향)

#### 출력
- 관계 거리 합이 최소인 학생 번호 한 개

#### 제한
- 입력은 보통 모든 학생이 친구 관계로 연결된 경우를 가정해도 된다.
- 시간 제한 2초, 메모리 제한 128MB

#### 예제 입력 1
```text
5 5
1 3
1 4
4 5
4 3
3 2
```

#### 예제 출력 1
```text
3
```

### [도전 기출문제] 보석 도둑
#### BOJ 1202

#### 문제
보석이 `N`개 있고, 각 보석은 무게 `Mi`와 가격 `Vi`를 갖는다. 가방이 `K`개 있으며, 가방 `i`에는 최대 무게 `Ci`까지 담을 수 있다. 가방 하나에는 보석을 최대 한 개만 넣을 수 있다. 훔칠 수 있는 보석 가격의 합의 최댓값을 구한다.

#### 입력
- 첫 줄: `N`, `K` (`1 <= N, K <= 300000`)
- 다음 `N`줄: `Mi`, `Vi` (`0 <= Mi, Vi <= 1000000`)
- 다음 `K`줄: `Ci` (`1 <= Ci <= 100000000`)

#### 출력
- 최대 가격 합 한 줄

#### 제한
- 시간 제한 1초, 메모리 제한 256MB

#### 예제 입력 1
```text
2 1
5 10
100 100
11
```

#### 예제 출력 1
```text
10
```

#### 예제 입력 2
```text

5 23

```

#### 예제 출력 2

````python
# Q2 TODO 코드 (학습자용)
# 아이디어: 가방을 작은 용량부터 쓰고, 담을 수 있는 보석 중 가치 최대를 고른다.

import heapq

n, k = map(int, input().split())
jewel = []
for _ in range(n):
    m, v = map(int, input().split())
    jewel.append((m, v))

bag = [int(input()) for _ in range(k)]

jewel.sort()
bag.sort()

pq = []
cnt = 0
ans = 0

for i in range(k):
    while cnt < n and None:
        # TODO: cnt번 보석의 무게가 현재 가방 용량 이하인 동안 힙에 넣기
        # 힌트: jewel cnt 무게와 bag i 비교
        heapq.heappush(pq, None)
        # TODO: 최대 힙처럼 쓰려면 heapq는 최소 힙이므로 가치에 부호를 어떻게 할지 정하기
        # 힌트: 가치 앞에 마이너스
        cnt += 1
    if pq:
        ans -= heapq.heappop(pq)

print(ans)
````

````python
# Q2 TODO 주석 포함 정답 코드
# 아이디어: 가방을 작은 용량부터 쓰고, 담을 수 있는 보석 중 가치 최대를 고른다.

import heapq

n, k = map(int, input().split())
jewel = []
for _ in range(n):
    m, v = map(int, input().split())
    jewel.append((m, v))

bag = [int(input()) for _ in range(k)]

jewel.sort()
bag.sort()

pq = []
cnt = 0
ans = 0

for i in range(k):
    while cnt < n and jewel[cnt][0] <= bag[i]:
        # TODO: cnt번 보석의 무게가 현재 가방 용량 이하인 동안 힙에 넣기
        # 힌트: jewel cnt 무게와 bag i 비교
        heapq.heappush(pq, -jewel[cnt][1])
        # TODO: 최대 힙처럼 쓰려면 heapq는 최소 힙이므로 가치에 부호를 어떻게 할지 정하기
        # 힌트: 가치 앞에 마이너스
        cnt += 1
    if pq:
        ans -= heapq.heappop(pq)

print(ans)
````

````python
import heapq

n, k = map(int, input().split())
jewel = []
for _ in range(n):
    m, v = map(int, input().split())
    jewel.append((m, v))

bag = [int(input()) for _ in range(k)]

jewel.sort()
bag.sort()

pq = []
cnt = 0
ans = 0

for i in range(k):
    while cnt < n and jewel[cnt][0] <= bag[i]:
        heapq.heappush(pq, -jewel[cnt][1])
        cnt += 1
    if pq:
        ans -= heapq.heappop(pq)

print(ans)
````

### 알고리즘 흐름도 (Q2)
```mermaid
flowchart TD

  R --> L[cnt는 0, pq 비움]
  L --> G[가방 하나 선택]

  H -->|아니오| Q{pq 비어 있지 않음}
  Q -->|예| V[pop 해서 ans에 가치 반영]
  Q -->|아니오| N[건너뜀]
  V --> M{가방 더 있음}
  N --> M
  M -->|예| G
  M -->|아니오| E[ans 출력]
```

### 가방 순서가 중요한 이유 (Q2)
```mermaid
flowchart TD
  A[작은 가방부터 보면] --> B[당장 넣을 수 있는 보석 집합이 점점 커짐]
  B --> C[큰 가방은 작은 가방이 넣을 수 있는 보석도 포함]
  C --> D[힙에 쌓인 후보는 이전 가방까지 고려된 상태]
  D --> E[그중 최대 가치를 빼 쓰면 탐욕이 최적]
```

**시간복잡도:** 정렬 `O(N log N + K log K)`, 힙 연산 amortized `O((N+K) log N)` 수준으로 1초 제한에 맞춘다.
