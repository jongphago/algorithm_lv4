````python
# Q1 간결 정답 코드
import sys
import heapq
input = sys.stdin.readline

n, m = map(int, input().split())
adj = [[] for _ in range(n + 1)]
ind = [0] * (n + 1)

for _ in range(m):
    a, b = map(int, input().split())
    adj[a].append(b)
    ind[b] += 1

hq = [i for i in range(1, n + 1) if ind[i] == 0]
heapq.heapify(hq)

res = []
while hq:
    cur = heapq.heappop(hq)
    res.append(cur)
    for nxt in adj[cur]:
        ind[nxt] -= 1
        if ind[nxt] == 0:
            heapq.heappush(hq, nxt)

print(*res)
````

````python
# Q2 간결 정답 코드
import sys
import heapq
input = sys.stdin.readline

N, M = map(int, input().split())
adj = [[] for _ in range(N + 1)]
ind = [0] * (N + 1)

for _ in range(M):
    data = list(map(int, input().split()))
    seq = data[1:]
    for a, b in zip(seq, seq[1:]):
        adj[a].append(b)
        ind[b] += 1

hq = [v for v in range(1, N + 1) if ind[v] == 0]
heapq.heapify(hq)

res = []
while hq:
    cur = heapq.heappop(hq)
    res.append(cur)
    for nxt in adj[cur]:
        ind[nxt] -= 1
        if ind[nxt] == 0:
            heapq.heappush(hq, nxt)

if len(res) != N:
    print(0)
else:
    sys.stdout.write("\n".join(map(str, res)))
````
