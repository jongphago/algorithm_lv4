````python
# Q1 TODO 주석 포함 정답 코드
# 아이디어: 플로이드-워셜로 모든 쌍 최단거리 계산 후 dist[i][i] 최소값 확인

INF = 4_000_001

V, E = map(int, input().split())
dist = [[INF] * (V + 1) for _ in range(V + 1)]

for _ in range(E):
    a, b, c = map(int, input().split())
    dist[a][b] = c

for k in range(1, V + 1):
    for i in range(1, V + 1):
        for j in range(1, V + 1):
            # TODO: i -> k -> j 경유가 더 짧으면 갱신
            # 힌트: dist[i][j]와 dist[i][k] + dist[k][j] 비교
            dist[i][j] = min(dist[i][j], dist[i][k] + dist[k][j])

min_cycle = INF
for i in range(1, V + 1):
    # TODO: dist[i][i]를 이용해 최소 사이클 길이 갱신
    # 힌트: min 함수 사용
    min_cycle = min(min_cycle, dist[i][i])

# TODO: 사이클이 없으면 -1, 있으면 최소값 출력
# 힌트: 조건식 (A if condition else B)
print(min_cycle if min_cycle != INF else -1)
````

````python
# Q1 간결 정답 코드
INF = 4_000_001
V, E = map(int, input().split())
dist = [[INF] * (V + 1) for _ in range(V + 1)]

for _ in range(E):
    a, b, c = map(int, input().split())
    dist[a][b] = c

for k in range(1, V + 1):
    for i in range(1, V + 1):
        dik = dist[i][k]
        for j in range(1, V + 1):
            if dist[i][j] > dik + dist[k][j]:
                dist[i][j] = dik + dist[k][j]

ans = min(dist[i][i] for i in range(1, V + 1))
print(ans if ans != INF else -1)
````

````python
# Q2 TODO 주석 포함 정답 코드
# 아이디어: 플로이드-워셜로 간접 비교 관계까지 채운 뒤, 각 학생의 비교 가능 인원 수 확인

N, M = map(int, input().split())

UNKNOWN = 2  # 아직 모름
YES = 1      # i < j 를 알 수 있음
SELF = 0     # 자기 자신

graph = [[UNKNOWN] * (N + 1) for _ in range(N + 1)]
for i in range(1, N + 1):
    graph[i][i] = SELF

for _ in range(M):
    a, b = map(int, input().split())
    graph[a][b] = YES

for k in range(1, N + 1):
    for i in range(1, N + 1):
        for j in range(1, N + 1):
            # TODO: i < k 이고 k < j 이면 i < j 확정
            # 힌트: graph[i][k], graph[k][j]를 확인
            if graph[i][k] == YES and graph[k][j] == YES:
                graph[i][j] = YES

result = 0
for i in range(1, N + 1):
    known = 0
    for j in range(1, N + 1):
        if i == j:
            continue
        # TODO: i와 j의 대소를 알 수 있으면 카운트 증가
        # 힌트: graph[i][j] == YES 또는 graph[j][i] == YES
        if graph[i][j] == YES or graph[j][i] == YES:
            known += 1

    # TODO: known이 N - 1이면 result 증가
    # 힌트: 모든 다른 학생과 비교 가능해야 등수 확정
    if known == N - 1:
        result += 1

print(result)
````

````python
# Q2 간결 정답 코드
N, M = map(int, input().split())

g = [[False] * (N + 1) for _ in range(N + 1)]
for _ in range(M):
    a, b = map(int, input().split())
    g[a][b] = True

for k in range(1, N + 1):
    for i in range(1, N + 1):
        if not g[i][k]:
            continue
        for j in range(1, N + 1):
            if g[k][j]:
                g[i][j] = True

ans = 0
for i in range(1, N + 1):
    known = 0
    for j in range(1, N + 1):
        if i != j and (g[i][j] or g[j][i]):
            known += 1
    if known == N - 1:
        ans += 1

print(ans)
````
