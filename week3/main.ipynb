## Q1
### [도전 기출문제] 경로 찾기
#### 방향 그래프 도달 가능성 (BOJ 11403)

## 문제

강의실을 정점으로, 일방통행 통로를 간선으로 생각한 방향 그래프가 주어진다.  
직접 연결뿐 아니라 여러 정점을 거쳐서도 이동할 수 있는지 확인해야 한다.

## 입력

- 첫째 줄: 정점(강의실) 수 `N` (`1 <= N <= 100`)
- 다음 `N`줄: 인접 행렬 정보
- `i`행 `j`열이 `1`이면 `i -> j`로 직접 이동 가능, `0`이면 불가능

## 출력

- 총 `N`줄 출력
- `i`행 `j`열은 `i`에서 `j`로 갈 수 있으면 `1`, 아니면 `0`
- 숫자는 공백으로 구분

## 제한

- 방향 그래프
- 자기 자신으로 가는 직접 간선 정보는 입력에서 `0`

### 시간제한

- 1초

### 메모리제한

- 256MB

## Q2
### [도전 기출문제] 플로이드
#### 모든 도시 쌍 최단 거리 (BOJ 11404)

## 문제

`N`개의 도시와 `M`개의 방향 도로 정보가 주어진다.  
한 도시에서 다른 도시로 가는 최소 비용을 모두 구해야 한다.

## 입력

- 첫째 줄: 도시 수 `N` (`1 <= N <= 100`)
- 둘째 줄: 도로 수 `M` (`1 <= M <= 10000`)
- 다음 `M`줄: `a b c` (`a -> b` 비용 `c`)
- 같은 `a, b`에 대해 여러 도로가 있을 수 있음

## 출력

- `N`줄 출력
- `i`행 `j`열: `i`에서 `j`로 가는 최소 비용
- 도달 불가능하면 `0` 출력

## 제한

- `1 <= c <= 100`
- 다중 간선은 더 짧은 비용만 반영

### 시간제한

- 1초

### 메모리제한

- 128MB

### 예제 입력 1

```text
5
14
1 2 2
1 3 3
1 4 1
1 5 10
2 4 2
3 4 1
3 5 1
4 5 3
3 5 10
3 1 8
1 4 2
5 1 7

5 2 4
```

### 예제 출력 1

```text

12 0 15 2 5
8 5 0 1 1
10 7 13 0 3
7 4 10 6 0
```

````python
# TODO 코드 (학습자용)
# 목표: 모든 도시 쌍의 최단 거리 구하기

INF = 10**9
N = int(input())
M = int(input())

dist = [[INF] * (N + 1) for _ in range(N + 1)]
for i in range(1, N + 1):
    dist[i][i] = 0

for _ in range(M):
    a, b, c = map(int, input().split())
    # TODO: 같은 a -> b 간선이 여러 개면 더 작은 비용 저장
    # 힌트: min(dist[a][b], c)
    dist[a][b] = None

for k in range(1, N + 1):
    for i in range(1, N + 1):
        for j in range(1, N + 1):
            # TODO: i -> k -> j 경로가 더 짧으면 갱신
            # 힌트: min(기존, 경유 비용)
            dist[i][j] = None

for i in range(1, N + 1):
    row = []
    for j in range(1, N + 1):
        # TODO: 도달 불가능(INF)면 0, 가능하면 거리값 추가
        if None:
            row.append(0)
        else:
            row.append(dist[i][j])
    print(*row)
````

````python
# TODO 주석 포함 정답 코드
# 목표: 모든 도시 쌍의 최단 거리 구하기

INF = 10**9
N = int(input())
M = int(input())

dist = [[INF] * (N + 1) for _ in range(N + 1)]
for i in range(1, N + 1):
    dist[i][i] = 0

for _ in range(M):
    a, b, c = map(int, input().split())
    # TODO: 같은 a -> b 간선이 여러 개면 더 작은 비용 저장
    # 힌트: min(dist[a][b], c)
    dist[a][b] = min(dist[a][b], c)

for k in range(1, N + 1):
    for i in range(1, N + 1):
        for j in range(1, N + 1):
            # TODO: i -> k -> j 경로가 더 짧으면 갱신
            # 힌트: min(기존, 경유 비용)
            dist[i][j] = min(dist[i][j], dist[i][k] + dist[k][j])

for i in range(1, N + 1):
    row = []
    for j in range(1, N + 1):
        # TODO: 도달 불가능(INF)면 0, 가능하면 거리값 추가
        if dist[i][j] == INF:
            row.append(0)
        else:
            row.append(dist[i][j])
    print(*row)
````

````python
# 간결 정답 코드 (제출용)
INF = 10**9
n = int(input())
m = int(input())

d = [[INF] * (n + 1) for _ in range(n + 1)]
for i in range(1, n + 1):
    d[i][i] = 0

for _ in range(m):
    a, b, c = map(int, input().split())
    d[a][b] = min(d[a][b], c)

for k in range(1, n + 1):
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            d[i][j] = min(d[i][j], d[i][k] + d[k][j])

for i in range(1, n + 1):
    row = [0 if d[i][j] == INF else d[i][j] for j in range(1, n + 1)]
    print(*row)
````

## Q3
### [도전 기출문제] 저울
#### 한국정보올림피아드 2014 지역본선

## 문제

서로 무게가 다른 `N`개 물건이 있다. 일부 물건 쌍에 대해 `A가 B보다 무겁다`는 결과가 주어진다.  
주어진 결과와 추론 가능한 결과를 이용해, 각 물건마다 비교 결과를 알 수 없는 물건 수를 구한다.

## 입력

- 첫째 줄: 물건 수 `N` (`5 <= N <= 100`)
- 둘째 줄: 비교 횟수 `M` (`0 <= M <= 2000`)
- 다음 `M`줄: `heavy light` (앞 물건이 더 무겁다)

## 출력

- `N`줄 출력
- `i`번째 줄: 물건 `i`와 비교 결과를 알 수 없는 물건의 개수

## 제한

- 모순되는 입력은 주어지지 않음

### 시간제한

- 1초

### 메모리제한

- 256MB

````python
# TODO 코드 (학습자용)
# 목표: 비교 불가능한 물건 개수 계산

N = int(input())
M = int(input())

graph = [[0] * (N + 1) for _ in range(N + 1)]

for _ in range(M):
    heavy, light = map(int, input().split())
    graph[heavy][light] = 1
    graph[light][heavy] = -1

for k in range(1, N + 1):
    for i in range(1, N + 1):
        for j in range(1, N + 1):
            # TODO: i > k 그리고 k > j 이면 i > j
            # 힌트: 1, 1 관계 전파
            if None:
                graph[i][j] = 1
                graph[j][i] = -1
            # TODO: i < k 그리고 k < j 이면 i < j
            # 힌트: -1, -1 관계 전파
            elif None:
                graph[i][j] = -1
                graph[j][i] = 1

for i in range(1, N + 1):
    unknown = 0
    for j in range(1, N + 1):
        # TODO: 자기 자신 제외, 비교 결과가 0이면 unknown 증가
        if None:
            unknown += 1
    print(unknown)
````

````python
# TODO 주석 포함 정답 코드
# 목표: 비교 불가능한 물건 개수 계산

N = int(input())
M = int(input())

graph = [[0] * (N + 1) for _ in range(N + 1)]

for _ in range(M):
    heavy, light = map(int, input().split())
    graph[heavy][light] = 1
    graph[light][heavy] = -1

for k in range(1, N + 1):
    for i in range(1, N + 1):
        for j in range(1, N + 1):
            # TODO: i > k 그리고 k > j 이면 i > j
            # 힌트: 1, 1 관계 전파
            if graph[i][k] == 1 and graph[k][j] == 1:
                graph[i][j] = 1
                graph[j][i] = -1
            # TODO: i < k 그리고 k < j 이면 i < j
            # 힌트: -1, -1 관계 전파
            elif graph[i][k] == -1 and graph[k][j] == -1:
                graph[i][j] = -1
                graph[j][i] = 1

for i in range(1, N + 1):
    unknown = 0
    for j in range(1, N + 1):
        # TODO: 자기 자신 제외, 비교 결과가 0이면 unknown 증가
        if i != j and graph[i][j] == 0:
            unknown += 1
    print(unknown)
````

````python
# 간결 정답 코드 (제출용)
n = int(input())
m = int(input())

g = [[0] * (n + 1) for _ in range(n + 1)]
for _ in range(m):
    a, b = map(int, input().split())
    g[a][b] = 1
    g[b][a] = -1

for k in range(1, n + 1):
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            if g[i][k] == 1 and g[k][j] == 1:
                g[i][j] = 1
                g[j][i] = -1
            elif g[i][k] == -1 and g[k][j] == -1:
                g[i][j] = -1
                g[j][i] = 1

for i in range(1, n + 1):
    print(sum(1 for j in range(1, n + 1) if i != j and g[i][j] == 0))
````
