````python
# 간결 정답 코드 (제출용)
import heapq
import sys

input = sys.stdin.readline

N = int(input())
heap = [int(input()) for _ in range(N)]
heapq.heapify(heap)

total = 0
while len(heap) > 1:
    merged = heapq.heappop(heap) + heapq.heappop(heap)
    total += merged
    heapq.heappush(heap, merged)

print(total)
````

````python
# 간결 정답 코드 (제출용)
import heapq
import sys

input = sys.stdin.readline

N, K = map(int, input().split())
customers = [tuple(map(int, input().split())) for _ in range(N)]

counter_heap = [(0, i) for i in range(K)]
heapq.heapify(counter_heap)

exit_heap = []
for customer_id, item_count in customers:
    current_time, counter_id = heapq.heappop(counter_heap)
    end_time = current_time + item_count
    heapq.heappush(counter_heap, (end_time, counter_id))
    heapq.heappush(exit_heap, (end_time, -counter_id, customer_id))

answer = 0
rank = 1
while exit_heap:
    _, _, customer_id = heapq.heappop(exit_heap)
    answer += rank * customer_id
    rank += 1

print(answer)
````
