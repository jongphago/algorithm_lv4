## 문제

디랩이는 자료구조 수업에서 **최소 힙(Min-Heap)** 을 배우고 있다.  
최소 힙은 **가장 작은 값을 빠르게 꺼낼 수 있는 자료구조**이다.

이해를 돕기 위해 디랩이는 아래와 같은 방식으로 최소 힙을 직접 실습하고 있다:

- 자연수가 주어지면, 그 수를 힙에 **추가**한다.
- **0**이 주어지면, 현재 힙에 들어 있는 수 중에서 **가장 작은 값을 꺼내고**, 그 값을 결과에 기록한다.
- 만약 힙이 **비어 있었다면**, 대신 0을 결과에 기록한다.
- 모든 명령이 주어진 후, 디랩이가 기록한 **모든 값의 합을 출력**하자.

---

## 입력

- 첫 줄에 명령의 개수 **N**이 주어진다. (1 ≤ N ≤ 100,000)
- 이어지는 N개의 줄에는 각각 하나의 정수가 주어진다.
    - 0보다 **큰 수가 주어지면 힙에 추가하는 명령**
    - **0이 주어지면 가장 작은 값을 꺼내는 명령**
    - 모든 정수는 2³¹보다 작은 자연수 또는 0이다.

---

## 출력

- 꺼낸 값들을 **모두 더한 합**을 한 줄에 출력하라.

---

### 시간제한

````python
import heapq

N = 18
h = []
nums = [1, -1, 0, 0, 0, 1, 1, -1, -1, 2, -2, 0, 0, 0, 0, 0, 0, 0]
for i in nums:
    if i == 0:
        heapq.heappush(h, abs(i))
    else:
        if not h:
            out = heapq.heappop(h)
        else:
            continue
print(h)
````

````python
import heapq

N = int(input())
h = []
total  = 0
for _ in range(N):
    num = int(input())
    
    if num == 0:
        if h:
            total += heapq.heappop(h)[-1]
        else:
            continue
    else:
        heapq.heappush(h, (abs(num), num))
print(total)
````
