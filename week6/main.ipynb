출발 `1`, 도착 `5`. `1→4` 비용 2, `4→5` 비용 2로 합계 4. 다른 경로는 모두 비용이 더 크다.

```mermaid
graph LR
    A[1] -->|2| B[4]
    B -->|2| C[5]
```
